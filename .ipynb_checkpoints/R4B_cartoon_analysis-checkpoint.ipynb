# A comprehensive analysis of representation of women in STEM in popular animated movies 

```python
# import libraries
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# import CSV file to a dataframe
filename = "movie_analysis.csv"
movies = pd.read_csv(filename, header=0)

movies
```

```python
contains_STEM_character = movies["female_STEM_character"] | movies["male_STEM_character"]
movies["contains_STEM_character"] = contains_STEM_character

movies.head()
```

```python
female_to_male_STEM_ratio = movies["num_female_STEM_characters"] / movies["num_male_STEM_characters"]
```

```python
movies["female_to_male_STEM_ratio"] = female_to_male_STEM_ratio

movies.head()
```

```python
STEM_movies = movies[movies["contains_STEM_character"]]

STEM_movies
```

## Average number of STEM characters

```python
# Average number of overall STEM characters
average_num_STEM_characters = np.mean(STEM_movies["num_female_STEM_characters"] + STEM_movies["num_male_STEM_characters"])
average_num_STEM_characters
```

```python
# Average number of male STEM characters
average_num_male_STEM_characters = np.mean(STEM_movies["num_male_STEM_characters"])
average_num_male_STEM_characters
```

```python
# Average number of female STEM characters
average_num_female_STEM_characters = np.mean(STEM_movies["num_female_STEM_characters"])
average_num_female_STEM_characters
```

## Average ratio of female:male STEM characters

```python
average_ratio = np.mean(STEM_movies["female_to_male_STEM_ratio"])
average_ratio
```

## Average probability of being a STEM protagonist

```python
# Simple probability of a movie having a STEM protagonist
prob_STEM_protagonist = np.mean(STEM_movies["is_STEM_protagonist"])
prob_STEM_protagonist
```

```python
# Simple probability of a movie having a female STEM protagonist
prob_female_STEM_protagonist = np.mean(STEM_movies["is_female_STEM_protagonist"])
prob_female_STEM_protagonist
```

```python
# Simple probability of a movie having a male STEM protagonist
prob_male_STEM_protagonist = np.mean(STEM_movies["is_male_STEM_protagonist"])
prob_male_STEM_protagonist
```

```python
# A probability of a protagonist being a man in STEM, given that the movie contains STEM protagonist
cond_prob_male_STEM_protagonist = prob_male_STEM_protagonist / prob_STEM_protagonist
cond_prob_male_STEM_protagonist
```

```python
# A probability of a protagonist being a woman in STEM, given that the movie contains STEM protagonist
cond_prob_female_STEM_protagonist = prob_female_STEM_protagonist / prob_STEM_protagonist
cond_prob_female_STEM_protagonist
```

# Portrayal of female STEM characters

```python
female_STEM_movies = STEM_movies[STEM_movies["female_STEM_character"]]
female_STEM_movies
```

## Percentage of movies depicting women in STEM as less attractive than non-STEM women

```python
female_STEM_attractiveness = np.average(female_STEM_movies["female_STEM_less_attractive"])
female_STEM_attractiveness
```

## Percentage of movies depicting women in STEM as less competent than men in STEM

```python
female_STEM_competence = np.average(female_STEM_movies["female_STEM_less_expert"])
female_STEM_competence
```
